```python
import re

def parse_prerequisite(prerequisite):
    # Step 1: Split by 'and' (lower precedence)
    and_parts = prerequisite.split('and')

    # Step 2: Split each part by 'or' (higher precedence)
    parsed_parts = []
    for part in and_parts:
        or_parts = [sub.strip() for sub in part.split('or')]
        parsed_parts.append(or_parts)

    # Step 3: Build the final structure
    return parsed_parts

def print_parsed_structure(parsed_structure):
    for i, group in enumerate(parsed_structure):
        if len(group) > 1:
            print(f"Group {i+1}: {' or '.join(group)}")
        else:
            print(f"Group {i+1}: {group[0]}")



# # Test with some examples
# example_1 = "A or B and C or D"
# example_2 = "A and B or C and D"
# example_3 = "A or B and C and D or E"
# example_4 = "A or B and C or D and E"

# # Parse the prerequisites
# parsed_1 = parse_prerequisite(example_1)
# parsed_2 = parse_prerequisite(example_2)
# parsed_3 = parse_prerequisite(example_3)
# parsed_4 = parse_prerequisite(example_4)

# # Print parsed structures
# print("Parsed Example 1:")
# print_parsed_structure(parsed_1)

# print("\nParsed Example 2:")
# print_parsed_structure(parsed_2)

# print("\nParsed Example 3:")
# print_parsed_structure(parsed_3)

# print("\nParsed Example 4:")
# print_parsed_structure(parsed_4)

# print(parsed_1)
# print(parsed_2)
# print(parsed_3)
# print(parsed_4)
```

```python
# Define regex patterns
course_pattern = r"(?P<course_code>[A-Z]{2,4} \d{3}[A-Z]?)\. (?P<course_name>.+?) \(\d\)"
prereq_pattern = r"Prerequisites: (?P<prerequisites>.+?)(?:[.;])"

# Extract course information
courses = {}
for match in re.finditer(course_pattern, course_text):
    course_code = match.group("course_code")
    course_name = match.group("course_name")
    courses[course_code] = {
        "course_name": course_name,
        "prerequisites": None
    }

# Extract prerequisites for each course
for course_code in courses.keys():
    course_section_pattern = rf"{course_code}\..+?Prerequisites: (?P<prerequisites>.+?)(?:[.;])"
    prereq_match = re.search(course_section_pattern, course_text, re.DOTALL)
    if prereq_match:
        prerequisites = prereq_match.group("prerequisites").strip()
        courses[course_code]["prerequisites"] = prerequisites

# Print the parsed course table
print("Course Table:")
for code, details in courses.items():
    print(f"{code}: {details['course_name']}")
    print(f"  Prerequisites: {details['prerequisites']}")



```
